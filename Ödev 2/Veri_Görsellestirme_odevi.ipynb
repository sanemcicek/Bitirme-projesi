## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  

R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar. 
State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
Profit (Kâr): Şirketin elde ettiği toplam kâr.  
Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='R&D Spend', y='Profit', s=100, alpha=0.7, color='blue')
plt.title('R&D Spend vs Profit', fontsize=16, fontweight='bold')
plt.xlabel('R&D Spend ($)', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('50_Startups.csv')

fig = px.scatter(data, x='R&D Spend', y='Profit', title='Ar-Ge Harcaması vs Kâr (Plotly)', size_max=10)
fig.update_layout(xaxis_title='Ar-Ge Harcaması ($)', yaxis_title='Kâr ($)')
fig.show()
```

## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıran bir çubuk grafik.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('50_Startups.csv')
avg_profit_by_state = data.groupby('State')['Profit'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_profit_by_state, x='State', y='Profit', palette='viridis')
plt.title('Average Profit by State', fontsize=16, fontweight='bold')
plt.xlabel('State', fontsize=12)
plt.ylabel('Average Profit ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('50_Startups.csv')
avg_profit_by_state = data.groupby('State')['Profit'].mean().reset_index()

fig = px.bar(avg_profit_by_state, x='State', y='Profit', title='Eyaletlere Göre Ortalama Kâr (Plotly)', color='State')
fig.update_layout(xaxis_title='Eyalet', yaxis_title='Ortalama Kâr ($)')
fig.show()
```

## 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının dağılımını karşılaştıran bir kutu grafiği.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('50_Startups.csv')
spend_data = data[['R&D Spend', 'Administration', 'Marketing Spend']].melt(var_name='Spend Type', value_name='Amount')
plt.figure(figsize=(10, 6))
sns.boxplot(data=spend_data, x='Spend Type', y='Amount', palette='Set2')
plt.title('Distribution of Spend Types', fontsize=16, fontweight='bold')
plt.xlabel('Spend Type', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv('50_Startups.csv')
spend_data = data[['R&D Spend', 'Administration', 'Marketing Spend']].melt(var_name='Harcama Türü', value_name='Miktar')

fig = px.box(spend_data, x='Harcama Türü', y='Miktar', title='Harcama Türlerinin Dağılımı (Plotly)', color='Harcama Türü')
fig.update_layout(xaxis_title='Harcama Türü', yaxis_title='Miktar ($)')
fig.show()
```
